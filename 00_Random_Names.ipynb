```python
# Cria uma lista com alguns estados Americanos
some_states = ['Alabama','Alaska','Arizona','Arkansas','California','Colorado','Connecticut','Florida','Georgia',
'New York','New Jersey','Virginia','Texas','Nevada']
```

```python
# Gera o primeiro dataframe contendo nomes aleatórios do sexo feminino distribuidos entre os 14 estados listados
import pandas as pd
import names 

df_female = pd.DataFrame()
for i in range(1,535):
    for j in range(0,14):
        df_female = df_female.append({'full_name': names.get_full_name(gender='female'),'gender':'F','state':some_states[j]}, ignore_index=True)

df_female
```

```python
# Gera o segundo dataframe contendo nomes aleatórios do sexo masculino distribuidos entre os 14 estados listados
df_male = pd.DataFrame()
for i in range(0,535):
    for j in range(0,14):
        df_male = df_male.append({'full_name': names.get_full_name(gender='male'),'gender':'M','state':some_states[j]}, ignore_index=True)
df_male
```

```python
# Concatenando os dataframes criados
person = pd.DataFrame()
person = pd.concat([df_male, df_female], axis=0)
```

```python
# Embaralhando os dados através do método sample do pandas
person = person.sample(frac=1).reset_index(drop=True)
person
```

```python
# Exportando o arquivo para .csv
person.to_csv('datasets/random_person.csv', index=False)
```
